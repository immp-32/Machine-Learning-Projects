# Making our Model Ready

```python
# independent variables
X = df[['area', 'bedrooms', 'bathrooms', 'stories']]

# dependent variables
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Creating a object of linear Regression model
linear_obj = LinearRegression()
```

```python

# now fitting the linear regression into the object
linear_obj.fit(X_train, y_train)
```

```python
predictions = linear_obj.predict(X_test)
```

# Plotting actual price vs Predicted Price

```python
plt.plot(np.arange(len(y_test)), y_test, label="Actual")
plt.plot(np.arange(len(predictions)), predictions, label="Predicted")

plt.legend()
plt.show()
```
